```python
from thicket import Thicket
tk = th.Thicket.from_caliperreader(glob("*.cali"))
tk.dataframe = tk.dataframe.drop(["nid","spot.channel","Total time","Min time/rank","Max time/rank"], axis = 1)
```

```python
tk.statsframe.dataframe["time"] = 1
tk.tree()
print(tk.tree())
```

```python
gb = tk.groupby("num_vals")
```

```python
ctk = th.Thicket.concat_thickets(
    thickets=list(gb.values()),
    headers=list(gb.keys()),
    axis="columns",
    metadata_key="num_threads"
)
```

```python
ctk.dataframe
```

```python
ctk.dataframe = ctk.dataframe.reset_index().drop(("node",""), axis = 1)


ctk.dataframe = ctk.dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])
```

```python
ctk.dataframe
```

```python
# Define common variables
function_names = set(ctk.dataframe.index.get_level_values("name"))
threads = sorted(list(set(ctk.dataframe.index.get_level_values("num_threads"))))
num_values = sorted(list(set(ctk.dataframe.columns.get_level_values(0))))
```

```python
def plot_mpl(df, title, xlabel, ylabel,legend):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(
        df,
        marker='o',
    )
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.legend(legend)
    plt.grid(True)
    plt.xticks(threads)
    plt.show()
    return fig, ax

# def plot_mpl(df, title, xlabel, ylabel):
#     fig = plt.figure()
#     ax = fig.add_subplot(1, 1, 1)
#     ax.plot(
#         df,
#         marker='o',
#     )
#     ax.set_title(title)
#     ax.set_xlabel(xlabel)
#     ax.set_ylabel(ylabel)
#     ax.legend(df.columns)
#     plt.grid(True)
#    # plt.yscale("log", base=10)
#   #  plt.xscale("log", base=2)
#     plt.xticks(threads)
#     plt.show()
#     return fig, ax
```
